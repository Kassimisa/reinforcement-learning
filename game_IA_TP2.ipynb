```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt

env = gym.make("FrozenLake-v1", is_slippery=True, render_mode=None)  # Désactiver le rendu graphique

num_states = env.observation_space.n
num_actions = env.action_space.n
q_table = np.zeros((num_states, num_actions))


def epsilon_greedy(q_table, state, epsilon):
    if np.random.rand() < epsilon:
        return env.action_space.sample()  # Exploration : action aléatoire
    else:
        return np.argmax(q_table[state])  # Exploitation : meilleure action connue

# Paramètres
alpha = 0.1  
gamma = 0.99  
epsilon = 1.0 
epsilon_decay = 0.995 
num_episodes = 5000  

# Pour stocker les récompenses cumulatives
rewards = []

# Boucle d’apprentissage
for episode in range(num_episodes):
    state, _ = env.reset()  # Extraire l'état du tuple retourné par reset()
    done = False
    total_reward = 0

    while not done:
        # ////////////////////////action///////////////
        action = epsilon_greedy(q_table, state, epsilon)

        next_state, reward, done, truncated, info = env.step(action)  

        #le nouveau q-table
        best_next_action = np.argmax(q_table[next_state])
        q_table[state, action] += alpha * (reward + gamma * q_table[next_state, best_next_action] - q_table[state, action])

        state = next_state
        total_reward += reward 

    
    epsilon *= epsilon_decay
    
    rewards.append(total_reward)  # Enregistrer la récompense totale de l'épisode

    if (episode + 1) % 1000 == 0:
        print(f"Épisode {episode + 1}, Récompense moyenne sur les 100 derniers épisodes : {np.mean(rewards[-100:])}")

# Afficher la Q-Table finale
print("Q-Table finale :")
print(q_table)


plt.plot(rewards)
plt.xlabel("Épisodes")
plt.ylabel("Récompense Cumulative")
plt.title("Évolution des récompenses avec Q-Learning")
plt.show()
```

```python
# Évaluation de l'agent après apprentissage
num_test_episodes = 100
successes = 0

eval_env = gym.make("FrozenLake-v1", is_slippery=True, render_mode="human")  # Activer le rendu

for _ in range(num_test_episodes):
    state, _ = eval_env.reset() 
    done = False

    while not done:
        action = np.argmax(q_table[state])
        next_state, reward, done, truncated, info = eval_env.step(action)  # Extraire next_state du tuple
        state = next_state

    if reward == 1:
        successes += 1

success_rate = successes / num_test_episodes
print(f"Taux de réussite : {success_rate * 100}%")

eval_env.close()
```
