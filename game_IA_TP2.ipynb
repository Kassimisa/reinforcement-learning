```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt

env = gym.make("FrozenLake-v1", is_slippery=True, render_mode=None)  # DÃ©sactiver le rendu graphique

num_states = env.observation_space.n
num_actions = env.action_space.n
q_table = np.zeros((num_states, num_actions))


def epsilon_greedy(q_table, state, epsilon):
    if np.random.rand() < epsilon:
        return env.action_space.sample()  # Exploration : action alÃ©atoire
    else:
        return np.argmax(q_table[state])  # Exploitation : meilleure action connue

# ParamÃ¨tres
alpha = 0.1  
gamma = 0.99  
epsilon = 1.0 
epsilon_decay = 0.995 
num_episodes = 5000  

# Pour stocker les rÃ©compenses cumulatives
rewards = []

# Boucle dâapprentissage
for episode in range(num_episodes):
    state, _ = env.reset()  # Extraire l'Ã©tat du tuple retournÃ© par reset()
    done = False
    total_reward = 0

    while not done:
        # ////////////////////////action///////////////
        action = epsilon_greedy(q_table, state, epsilon)

        next_state, reward, done, truncated, info = env.step(action)  

        #le nouveau q-table
        best_next_action = np.argmax(q_table[next_state])
        q_table[state, action] += alpha * (reward + gamma * q_table[next_state, best_next_action] - q_table[state, action])

        state = next_state
        total_reward += reward 

    
    epsilon *= epsilon_decay
    
    rewards.append(total_reward)  # Enregistrer la rÃ©compense totale de l'Ã©pisode

    if (episode + 1) % 1000 == 0:
        print(f"Ãpisode {episode + 1}, RÃ©compense moyenne sur les 100 derniers Ã©pisodes : {np.mean(rewards[-100:])}")

# Afficher la Q-Table finale
print("Q-Table finale :")
print(q_table)


plt.plot(rewards)
plt.xlabel("Ãpisodes")
plt.ylabel("RÃ©compense Cumulative")
plt.title("Ãvolution des rÃ©compenses avec Q-Learning")
plt.show()
```
